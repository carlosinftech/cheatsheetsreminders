```python
df_combine_a = pd.DataFrame(
    {"character":['A','B','C'],
     "number":['1','2','3']
})
df_combine_b = pd.DataFrame(
    {"character":['A','B','D'],
     "boolean":['0','1','0']
})
```

```python
pd.merge(df_combine_a, df_combine_b,
how='left', on='character')
#left Join matching rows
```

```python
pd.merge(df_combine_a, df_combine_b,
how='right', on='character')
#right Join matching rows
```

```python
pd.merge(df_combine_a, df_combine_b,
how='inner', on='character')
#Inner Join matching rows
```

```python
pd.merge(df_combine_a, df_combine_b,
how='outer', on='character')
#Outer Join matching rows
```
